<div style="background-color: lightgrey; color: green">Introduction:</div>
Polaflow is a package that calculate the numeric propagation of an initial state using a Shao Wan method modified, such that it can calculate with non-linear term, life time term and coherent pumping term, furthermore it contains auxiliarity functions that facilitate the data analysis.

<div style="background-color: lightgrey; color: green">Instalation:</div>
For use Polaflow is nescessary to have python2.7.x, NumPy 1.8.x, SciPy 0.13.x, Matplotlib 1.3.x and Ipython2.1.x [Notebook] installed  or later. After this the files ShaoWangV3.py and Ploaflow.py must be copied in the work directory.

<div style="background-color: lightgrey; color: green">the code use:</div><br>
For use of polaflow code is necessary load from Polaflow package the objets Particles and Polaflow1D, next define the particles to be calculates:<br>
part1=Particle()<br>
part2=Particle()<br>
part3=Polaflow1D(part1,part2), is important the order, because the first particle in Polaflow objet is the cavity and the second one is the exciton (it has the non-linearity).
<br><br>
<i>Porperties of the class Particle()</i>:<br>
obj.name: string<br>
obj.mass: float<br>
obj.w0: float<br>
obj.gamma: float<br>
obj.g: float<br>
obj.Psi0: numpy array (complex or real float)<br>
obj.Potential: numpy array (complex or real float)<br>
obj.Psi: numpy array (complex float)<br>
<br>
<i>Methods of the class Particle()</i>:<br>
obj.Gsummary(): this present a summary of the initial data of the particle<br>
obj.GraphXT(): this method graphs the time evolution of the spatial wave density<br>
<br>
<i>Porperties of the class Polaflow1D()</i>:<br>
obj.name: string<br>
obj.x : numpy array (real float)<br>
obj.t : numpy array (real float)<br>
obj.M : int<br>
obj.R : int <br>
obj.hbar : float<br>
obj.rabi : float  <br>              
obj.pathwd : string with path structure<br>
obj.path : string with path structure<br>
obj.tolerance : float<br>
obj.sweeps : int<br><br>

<i>Methods of the class Polaflow1D()</i>:<br>
obj.error: it shows the error acumulated in the calculus<br>
obj.Refres(): it refresh the changes done in the data and constants<br>
obj.Save(): it saves the data in the path defined by obj.pathwd, if you are no sure please don change this value<br>
obj.Load(): it loads the data present in some directory defined by obj.path<br>
obj.Calculate(): it runs the Shao Wang method<br>
obj.PlotKE(): it does the dispersion plot<br>

<div style="color: red">Example 1:</div>
Rabi free oscillation

Defining particles:

```python
from Polaflow import Particle,Polaflow1D
photon=Particle()
exciton=Particle()
polariton=Polaflow1D(photon,exciton)
```

It is defined the functions that will be used, these can have any spacial structure

```python
def gauss(x,sigma=20):
    from numpy import exp, sqrt, pi                                                     
    return  exp(-0.5*x**2/sigma**2)/sqrt(sqrt(pi*sigma**2))
```

The constants and values are initialized and is finished with obj.Refresh()

```python
from numpy import zeros,linspace
J=2**6
x=linspace(-100,100,J)
t=linspace(0,20,2**6)
photon.Psi0=gauss(x,sigma=20.0)
exciton.Psi0=zeros(J)


photon.name="Photon"
photon.mass=0.025
photon.gamma=0.0
photon.Potential=zeros(J)

exciton.name="Exciton"
exciton.mass=2
exciton.gamma=0.0
exciton.g=0
exciton.Potential=zeros(J)

polariton.x=x
polariton.t=t
polariton.rabi=2.62989731
polariton.hbar=0.65731
polariton.M=10
polariton.R=10
polariton.name="Polaflow"
polariton.Refresh()
```

It  calculates the temporal propagation

```python
polariton.Calculate()
```

Save the data

```python
polariton.Save()
```

Space-time graphs

```python
%matplotlib inline
photon.PlotXT(contrast=2);
exciton.PlotXT();
```

Energy-dispersion graph

```python
%matplotlib inline
polariton.PlotKE(contrast=2);
```

```python
polariton.path="DataPF/Polaflow_2014_08_05_16_55_37"
polariton.Load()
```

<div style="color: red">example 2:</div>
    
Soliton

```python
def soliton1D(x,t,asol=0.4,vsol=0.0,x0sol=0.0):
    from numpy import exp,cosh
    return asol*exp(1.0j*(vsol*x+(asol**2-vsol**2)*t/2))/cosh(asol*(x-vsol*t-x0sol))


from Polaflow import Particle,Polaflow1D
from numpy import zeros,linspace
cero=Particle()
sol=Particle()
soliton=Polaflow1D(cero,sol)


J=2**8
x=linspace(-30,70,J)
t=linspace(0,10,2**7)
sol.Psi0=soliton1D(x,0,vsol=3.0)
cero.Psi0=zeros(J)


cero.name="cero"
cero.mass=1.0
cero.Potential=zeros(J)

sol.name="soliton"
sol.mass=1
sol.g=-1
sol.Potential=zeros(J)

soliton.x=x
soliton.t=t
soliton.rabi=0
soliton.hbar=1
soliton.M=10
soliton.R=10
soliton.name="Soliton"
soliton.Refresh()
```

```python
%matplotlib inline
cero.Gsummary();
sol.Gsummary();
```

```python
soliton.Calculate()
```

```python
%matplotlib inline
sol.PlotXT(contrast=1);
```

```python
%matplotlib inline
soliton.PlotKE(contrast=1);
```

Individual manipulation

```python
%matplotlib inline
import matplotlib.pyplot as plt
n=70
fig, axes = plt.subplots(1,1,figsize=(16,6.87))
axes.plot(sol.x,soliton1D(sol.x,sol.t[n],vsol=3.0).real,"b")
axes.plot(sol.x,sol.Psi[n].real,"r.");
```
