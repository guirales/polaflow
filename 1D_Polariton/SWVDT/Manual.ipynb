<div style="background-color: lightgrey; color: green">Introducción:</div>
Polaflow es un paquete que calcula la propagación numérica de un estado en un tiempo inicial usando el método propuesto por Shao Wan modificado para que calcule con terminos de no linealidad, decaimiento y bombeo, además de que contiene funciones auxiliares que facilitan el análisis y el manejo de los datos obtenidos.

<div style="background-color: lightgrey; color: green">Instalación:</div>
Para usar Polaflow es necesario tener instalado python2.7.x,  NumPy 1.8.x, SciPy 0.13.x, Matplotlib 1.3.x y Ipython2.1.x [Notebook] o superior. Luego de cumplir los prerequisitos se deben poner los archivos ShaoWangV3.py y Ploaflow.py en el que será el directorio de trabajo.


<div style="color: red">Ejemplo1:</div>
Oscilación libre

Definiendo particulas:

```python
from Polaflow import Particle,Polaflow1D
foton=Particle()
exciton=Particle()
polariton=Polaflow1D(foton,exciton)
```

Se define las funciones que se usaran, estas pueden tener cualquier estructura espacial

```python
def gauss(x,sigma=20):
    from numpy import exp, sqrt, pi                                                     
    return  exp(-0.5*x**2/sigma**2)/sqrt(sqrt(pi*sigma**2))
```

Se inicializan los valores y se finaliza con Object.Refres()

```python
from numpy import zeros,linspace
J=2**7
x=linspace(-100,100,J)
t=linspace(0,40,2**7)
foton.Psi0=gauss(x,sigma=1.0)
exciton.Psi0=zeros(J)


foton.name="Foton"
foton.mass=0.025
foton.Potential=zeros(J)

exciton.name="Exciton"
exciton.mass=2
exciton.Potential=zeros(J)

polariton.x=x
polariton.t=t
polariton.rabi=2.62989731
polariton.hbar=0.65731
polariton.M=10
polariton.R=10
polariton.name="Polaflow"
polariton.Refresh()
```

Se puede verificar los datos de entrada

```python
%matplotlib inline
foton.Gsummary();
exciton.Gsummary();
```

Para calcular la propagación del estado inicial

```python
polariton.Calculate()
```

Para guardar los datos generados

```python
polariton.Save()
```

Graficos de salida:
Para un gráfico de posición y tiempo

```python
%matplotlib inline
foton.PlotXT(contrast=1);
exciton.PlotXT();
```

Para un gráfico de dispersion y energía

```python
%matplotlib inline
polariton.PlotKE(contrast=2);
```

<div style="color: red">ejemplo 2:</div>
    
Soliton

```python
def soliton1D(x,t,asol=0.4,vsol=0.0,x0sol=0.0):
    from numpy import exp,cosh
    return asol*exp(1.0j*(vsol*x+(asol**2-vsol**2)*t/2))/cosh(asol*(x-vsol*t-x0sol))


from Polaflow import Particle,Polaflow1D
from numpy import zeros,linspace
cero=Particle()
sol=Particle()
soliton=Polaflow1D(cero,sol)


J=2**7
x=linspace(-10,50,J)
t=linspace(0,10,2**7)
sol.Psi0=soliton1D(x,0,vsol=3.0)
cero.Psi0=zeros(J)


cero.name="cero"
cero.mass=1.0
cero.Potential=zeros(J)

sol.name="soliton"
sol.mass=1
sol.g=-1
sol.Potential=zeros(J)

soliton.x=x
soliton.t=t
soliton.rabi=0
soliton.hbar=1
soliton.M=10
soliton.R=10
soliton.name="Soliton"
soliton.Refresh()
```

```python
%matplotlib inline
cero.Gsummary();
sol.Gsummary();
```

```python
soliton.Calculate()
```

```python
%matplotlib inline
sol.PlotXT(contrast=1);
```

```python
%matplotlib inline
soliton.PlotKE(contrast=1);
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
n=70
fig, axes = plt.subplots(1,1,figsize=(16,6.87))
axes.plot(sol.x,soliton1D(sol.x,sol.t[n],vsol=3.0).real,"b")
axes.plot(sol.x,sol.Psi[n].real,"r.");
```
